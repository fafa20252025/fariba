```python
def remove_outliers(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]
    return data
```

```python
data_without_outliers = remove_outliers(data, data.columns)
```

# **Machine learning process**

```python
X=data.drop('medv',axis=1)
y=data['medv']
```

```python
#split the data into training ans testing sets
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=100)   #train_test_split is function. test_size is 20%.
```

# **#apply linear Regression on training data**

```python
Lin_reg=LinearRegression() #Make a class and make an object of the class
Lin_reg.fit(x_train,y_train)  #training process starts here.  BEST FIT LINE or ML model will be stored in linear_reg object. FIT is a function
```

```python
Lin_reg.coef_ #the values are m1 to m13 because there are 13 columuns
```

```python
Lin_reg.intercept_ #c value
```

```python
Lin_reg.predict(x_test)
```

```python
y_pred=Lin_reg.predict(x_test)# we strore the above results in y_pred variable.
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':y_pred}) ## we compare the actual prediction to predicted values
```

# **performance metrics used in regression**

1- R2 score (r2_squared)
2-adjusted R2 Score
3-mean squared error (mse)
4- root mean squarred error (Rmse)

```python
r2_score(y_test,y_pred) #overall comparison . r2_score is a function
```

```python
1-(1-r2_score(y_test,y_pred))*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1) #adjusted r2_score
```

```python
mse=mean_squared_error(y_test,y_pred) #means squarred error. The lower mse menas the better our model is.
mse
```

```python
rmse=np.sqrt(mse) #root means squarred error
rmse
```

# **checking VIF score for muilicollinearity**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

VIF (variance inflation factor) is the correlation between two columns. if VIF >5 then two coloumns are highly corelated.

```python
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_train.values,i) for i in range(x_train.shape[1])] #we create a column named VIF.
#then we apply varican_inflation_factor on x_train.values on each row.
vif['Features']=x_train.columns #we create a column called features.
vif ['vif']=round(vif['vif'],2)
vif=vif.sort_values(by='vif',ascending=False)
vif
```
