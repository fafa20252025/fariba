```python
Q1 = heart[['trestbps', 'chol', 'thalach', 'oldpeak']].quantile(0.25)
Q3 = heart[['trestbps', 'chol', 'thalach', 'oldpeak']].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
heart = heart[~((heart[['trestbps', 'chol', 'thalach', 'oldpeak']] < lower_bound) | (heart[['trestbps', 'chol', 'thalach', 'oldpeak']] > upper_bound)).any(axis=1)]

print(f"DataFrame shape after outlier removal: {heart.shape}")
```

### Outlier Removal using IQR Method
To remove outliers, we use the Interquartile Range (IQR) method. This method identifies values that fall outside of 1.5 times the IQR from the first (Q1) and third (Q3) quartiles.

1.  **Calculate Q1 and Q3**: The 25th and 75th percentiles of the specified columns.
2.  **Calculate IQR**: The difference between Q3 and Q1.
3.  **Determine Bounds**: Any value below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR` is considered an outlier.
4.  **Filter Data**: Rows containing outliers in the specified columns are removed from the DataFrame.

```python
Q1 = heart[['trestbps', 'chol', 'thalach', 'oldpeak']].quantile(0.25)
Q3 = heart[['trestbps', 'chol', 'thalach', 'oldpeak']].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
heart = heart[~((heart[['trestbps', 'chol', 'thalach', 'oldpeak']] < lower_bound) | (heart[['trestbps', 'chol', 'thalach', 'oldpeak']] > upper_bound)).any(axis=1)]

print(f"DataFrame shape after outlier removal: {heart.shape}")
```

# **###Machine Learning process**

```python
# create variables (x and y)
x = heart.drop('target', axis=1)  #it means drop target from the columns and print every column
y = heart['target']
```

```python
# split data into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

**##standardisation /scaling of the data**



```python
scaler=StandardScaler()  #standardscaler is a class so we need to make an object of class (scaler).
x_train_scaled=scaler.fit_transform(x_train) # we take x_train data and store them in x_train_scaled variable
x_test_scaled=scaler.transform(x_test)
```

##applying Logistic Regression on the data

```python
log_reg=LogisticRegression()  #LogisticRegression is a class and we make an object of it (log_reg)
log_reg.fit(x_train_scaled,y_train)
```

```python
y_pred=log_reg.predict(x_test_scaled)
```
