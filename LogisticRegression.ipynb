### Outlier Removal using IQR Method
To remove outliers, we use the Interquartile Range (IQR) method. This method identifies values that fall outside of 1.5 times the IQR from the first (Q1) and third (Q3) quartiles.

1.  **Calculate Q1 and Q3**: The 25th and 75th percentiles of the specified columns.
2.  **Calculate IQR**: The difference between Q3 and Q1.
3.  **Determine Bounds**: Any value below `Q1 - 1.5 * IQR` or above `Q3 + 1.5 * IQR` is considered an outlier.
4.  **Filter Data**: Rows containing outliers in the specified columns are removed from the DataFrame.

```python
Q1 = heart[['trestbps', 'chol', 'thalach', 'oldpeak']].quantile(0.25)
Q3 = heart[['trestbps', 'chol', 'thalach', 'oldpeak']].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
heart = heart[~((heart[['trestbps', 'chol', 'thalach', 'oldpeak']] < lower_bound) | (heart[['trestbps', 'chol', 'thalach', 'oldpeak']] > upper_bound)).any(axis=1)]

print(f"DataFrame shape after outlier removal: {heart.shape}")
```
