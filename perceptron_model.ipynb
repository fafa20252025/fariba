```python
x1=[10,8,15] #dataset
x2=[8,20,25] #dataset
y=[20,15,30] #y is target value    #y=w1x1+w2x2
n=len(x1)
alpha=0.0001
```

Find best values of w1 , w2 and b that fit above dataset.
Equation is y=w1x1+w2x2+b

solution:
we use Gradient descent algorithum to solve the problem.
1- we randomly choose w1 and w2 value

```python
import numpy as np
np.random.seed(123) #seed always return same values when random is picked
random_init=np.random.randn(3) #3 means 3 values for w
w1,w2,b=random_init[0],random_init[1],random_init[2]
print(w1,w2,b)
```

2- compute MSE
MSE=predicted Y -actual Y / n =(w1x1+w2x2=b)**2 /n

```python
def MSE(w1,w2,b):

  y_pred=[]
  total_error=0

  for i in range(n):
    y_hat=w1*x1[i]+w2*x2[i]+b
    y_pred.append(y_hat)
    error=(y[i]-y_hat)**2  #error=(predicted value-actual value) to the power of 2
    total_error+=error
  return y_pred,total_error/n
```

```python
y_pred  #compare the predicted value to actual values of dataset. There are error in prediction
```

```python
def weight_update(w1, w2, b, y_pred):
  for i in range(n):
    dmse_dw1=2*(y_pred[i]-y[i])*x1[i]
    dmse_dw2=2*(y_pred[i]-y[i])*x2[i]
    dmse_db=2*(y_pred[i]-y[i])

    w1=w1-alpha*dmse_dw1
    w2=w2-alpha*dmse_dw2
    b=b-alpha*dmse_db
  return w1,w2,b
```

```python
from logging import error

import numpy as np
##parameter defenition
alpha=0.00065
#
#weight initialisation
np.random.seed(123)
random_init=np.random.randn(3)
w1,w2,b=random_init[0],random_init[1],random_init[2]
print(w1,w2,b)

#orchestration
n=len(x1)
iteration=100
error=[]
for inter_num in range(iteration):
  y_pred,total_error=MSE(w1,w2,b)

  print(total_error)
  error.append(total_error)
  w1,w2,b=weight_update(w1,w2,b,y_pred)
```

```python
#iteration1
y_pred,total_error=MSE(w1,w2,b)
print(total_error)

w1,w2,b=weight_update(w1,w2,b,y_pred)
print(w1,w2,b)

#iteration2
y_pred,total_error=MSE(w1,w2,b)
print(total_error)

w1,w2,b=weight_update(w1,w2,b,y_pred)
print(w1,w2,b)

#iteration3
y_pred,total_error=MSE(w1,w2,b)
print(total_error)

w1,w2,b=weight_update(w1,w2,b,y_pred)
print(w1,w2,b)


#the error keep increasing . now inorder to decrease the erro we need to change alpha value to smaller value from 0.01 to 0.001 or futher to 0.0001
```
