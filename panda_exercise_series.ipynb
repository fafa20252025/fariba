```python
import pandas as pd
```

# **Indexing**

```python
s[2:4]
```

# **iloc**
is a function used for location based indexing

```python
s.iloc[3]
```

```python
s.iloc[[1,3,4]]
```

```python
Index=['apple','banana','cherry','date','elderberry']  #index is the name of the variable
s.index=Index
s  #print s
```

```python
s['banana']
```

# **loc**
is a function for label based indexing. note that start and stop index are both included in the output unlike numerical indexing.

```python
s.loc['banana']
```

```python
s.loc['apple':'cherry']
```

```python
s.loc[['apple','cherry','elderberry']]  #multiple indexing
```

# **Dictionary**

```python
fruit_dict={'apple':10,'banana':20,'cherry':30,'date':40,'elderberry':50}
```

```python
#create series of dictionary
s2=pd.Series(fruit_dict, name='protein')
s2
```

# **Conditional selection**
example: get all the fruit that have protein more than 20

```python
s2[s2>20]
```

# **Logical operators**
AND, OR,NOT


```python
#fruits with protein over 20 and less than 50
s2[(s2>20) & (s2<=50)]
```

# *Modifying the series*

```python
#modify banana protein from 20 to 60
s2['banana']=60
s2
```
