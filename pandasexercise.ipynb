# **##Creating series using a dictionary**

```python
fruit_protein = {
    "avodado" :2.3,
    "banana"  :1.2,
    "apple"   :1.3,
    "kiwi"    :2.1,
    "mango"   :2.4,
    "cherries":2.4
}
```

```python
s2[s2>2]
```

# **#Data cleaning**
two ways
1- drop the rows with null values
2- replace null values with some value (the most repeated values in the column)

```python
# first we check null values
df.isnull().sum()
```

```python
# 1- first way : drop null values
df.dropna() #.This drops the null rows and the original dataframe is untouched.
```

```python
#2 -second way: replace null value
df['age'].fillna(df['age'].mean(), inplace=True)
df
```

```python
df[['salary', 'promoted salary']] = df[['salary', 'promoted salary']].fillna(df[['salary', 'promoted salary']].median())
df
```

```python
# REPLACE a value
df["name"]=df["name"].replace("alice", "sharon")   #This will store the change in original dataframe
df
```

```python
#Duplicated values
df_dup=df[df.duplicated(keep="first")]  #keep=first is the parameter for the function duplicate and it keeps the first duplicated row
df_dup
```

```python
# invalie values:
#Lambda

df["promoted salary"]=df["promoted salary"].apply(lambda x: x/10 if x>30000 else x)   #x is input value and the out put valude is devided by 10.
df
```

# Apply Method and lambda function
exp: multiply age by 2

```python

#multiply age by 2
def multiply_age(x):
  return x*2

df["age"]= df['age'].apply(multiply_age)
df
```

```python
# bring back the age to the original figures
df["age"]= df["age"].apply(lambda x: x/2)
df
```
