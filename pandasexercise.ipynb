# **##Creating series using a dictionary**

**1- create a dictionary**

```python
fruit_protein = {
    "avodado" :2.3,
    "banana"  :1.2,
    "apple"   :1.3,
    "kiwi"    :2.1,
    "mango"   :2.4,
    "cherries":2.4
}
```

**2- create a series of the above dictionary**

```python
s2=pd.Series(fruit_protein, name="protein")
s2  #print series
```

# **Conditional selection**
exp: show all fruits that have protein over 2 grams
(Boolean indexing)

```python
s2[s2>2]
```

# **Logical operators**
True and True =True
Truen and False =False
False and False=False
True or True= True
True or False =True
False or Flase=Flse


```python
# show fruits which have greater than 1 but have less than 2 gram protein.

s2[(s2>1)&(s2<2)]
```

```python
#not operation  not freater than 1.2
s2[~(s2>1.2)]
```

# **Modifying the series**


```python
#modify the protein of a fruit
s2["banana"]=3
s2
```
