# **##Creating series using a dictionary**

**1- create a dictionary**

```python
fruit_protein = {
    "avodado" :2.3,
    "banana"  :1.2,
    "apple"   :1.3,
    "kiwi"    :2.1,
    "mango"   :2.4,
    "cherries":2.4
}
```

**2- create a series of the above dictionary**

```python
s2=pd.Series(fruit_protein, name="protein")
s2
```

# **Conditional selection**
exp: show all fruits that have protein over 2 grams
(Boolean indexing)

```python
s2[s2>2]
```
